```python
from pathlib import Path

from mamba_agents import Agent, AgentConfig, AgentSettings
from mamba_agents.skills import SkillConfig
from mamba_agents.tools import glob_search, grep_search, run_bash
from mamba_agents.tools.filesystem import (
    append_file,
    copy_file,
    delete_file,
    file_info,
    list_directory,
    move_file,
    read_file,
    write_file,
)

tools = [
    read_file,
    write_file,
    append_file,
    list_directory,
    file_info,
    delete_file,
    move_file,
    copy_file,
    run_bash,
    glob_search,
    grep_search,
]


config = AgentConfig()

settings = AgentSettings()

agent = Agent(
    config=config,
    settings=settings,
    skill_dirs=[Path("skills")],
    tools=tools,
)

agent.list_skills()
```

```python
agent.invoke_skill("data-analyzer", "example.csv")
```

```python
result = await agent.run("Use the data-analyzer skill to analyze the file example.csv")

print(result.output)
```
